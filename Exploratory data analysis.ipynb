```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
```

### Read Data into dataframe

```python
data_20170101_20181130 = pd.read_csv('./datasets/TaiPower_20170101_20181130.csv')
data_20180101_20190131 = pd.read_csv('./datasets/TaiPower_20180101_20190131.csv')
```

### Check for Null values

```python
print('data_20170101_20181130 Null Values')
for column in data_20170101_20181130:
    if data_20170101_20181130[column].isnull().any():
       print('{0} has {1} null values'.format(column, data_20170101_20181130[column].isnull().sum()))
```

```python
print('data_20180101_20190131 Null Values')
for column in data_20180101_20190131:
    if data_20180101_20190131[column].isnull().any():
       print('{0} has {1} null values'.format(column, data_20180101_20190131[column].isnull().sum()))
```

## Seperate 2017 data

```python
data_2017 = data_20170101_20181130[data_20170101_20181130['日期'] <= 20171231]
```

## Seperate 2018 data

```python
data_2018 = data_20180101_20190131[data_20180101_20190131['日期'] <= 20181231]
```

### A peek of 2018 data

```python
data_2018.head()
```

```python
def plot_year(dates, value): 
    fig = plt.figure(figsize=(20,4))
    ax = fig.add_subplot(111)
    x_axis = pd.to_datetime(dates, format='%Y%m%d')
    ax.plot_date(x_axis, value, '-')
    fig.show()
```

```python
plot_year(data_2017['日期'], data_2017['淨尖峰供電能力(MW)'])
```

```python
plot_year(data_2018['日期'], data_2018['淨尖峰供電能力(MW)'])
```
